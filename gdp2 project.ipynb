```python
import pandas as pd
df=pd.read_excel("C:/Users/Hp/Downloads/GDP Data.xlsx")
df
```

```python
df2=pd.melt(df,id_vars=['Country Name','Country Code','Series Name','Series Code'],var_name='years',value_name='gdp')
df2.drop(columns=['Series Code'],inplace=True)
df2
```

```python
df3=pd.pivot_table(df2,index=['Country Name','Country Code','years'],columns='Series Name',values='gdp',
                   aggfunc='sum')
df3.reset_index(inplace=True)
gdp=df3
gdp
```

```python
gdp.info()
```

```python
gdp[['year', 'year2']]=gdp['years'].str.split('[', 1, expand=True)
gdp
```

```python
gdp.drop(columns=['years', 'year2'], inplace=True)
gdp
```

```python
gdp.rename(columns={'GDP (current US$)':'GDP','GDP per capita (current US$)':'GDP per capita'},inplace=True)
gdp
```

```python
gdp['GDP'] = pd.to_numeric(gdp['GDP'], errors='coerce')
```

```python
gdp['GDP per capita'] = pd.to_numeric(gdp['GDP per capita'], errors='coerce')
```

```python
gdp['year'] = gdp['year'].astype(int)
```

```python
gdp.dtypes
```

```python
gdp.head(30)
```

```python
gdp.isnull().sum()
```

```python
gdp.fillna(value={'GDP':0,'GDP per capita':0},inplace=True)
gdp
```

```python
gdp.isnull().sum()
```

```python
gdp.drop(columns=['Country Code'],inplace=True)
gdp
```

# Visualisation

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = total.groupby('year').agg({'GDP per capita': 'mean', 'hdi': 'mean', 'no of suicide': 'sum'})
a
```

```python
average_gdp=total.groupby(['Region Name'])['GDP per capita'].mean().sort_values(ascending=True)
average_gdp
```

```python
average_gdp.plot(kind='bar', title='region wise average gdp_pc',
               ylabel='Gdp per capita', xlabel='Region name', figsize=(10, 6))
```

```python
average_hdi=total.groupby(['Region Name'])['hdi'].mean().sort_values(ascending=True)
average_hdi
```

```python
average_hdi.plot(kind='bar', title='region wise average hdi',
               ylabel='hdi', xlabel='region name', figsize=(10, 6))
```

```python
suicide_by_age=total.groupby(['age range'])['no of suicide'].sum().sort_values(ascending=False)
suicide_by_age
```

```python
suicide_by_age.plot(kind='bar', title='age wise no of suicide',
               ylabel='no of suicide', xlabel='age range', figsize=(10, 6))
```

```python
hdi_rank_2021=total.groupby(['Country Name'])['hdi_rank_2021'].mean()
a=hdi_rank_2021.sort_values(ascending=True).head(10)
```

```python
a.plot(kind='bar',title='hdi rank 2021', xlabel='country', ylabel='rank',figsize=(10,6))
```
